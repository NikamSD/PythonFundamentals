```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    def __init__(self, maker):
        self.maker = maker
    
    def drivetype(self):
        print("This vehicle is designed for road travel.")
    
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):

    def __init__(self, maker, model):
        super().__init__(maker)
        self.model = model

    def start_engine(self):
        return f"The engine of the {self.maker} {self.model} is starting."

class Motorcycle(Vehicle):

    def __init__(self, maker, type_):
        super().__init__(maker)
        self.type_ = type_

    def start_engine(self):
        return f"The engine of the {self.maker} {self.type_}  is starting."


def vehicle_info(vehicle):
    print(vehicle.start_engine())
    vehicle.drivetype()
```

```python
car = Car("BMW", "X5")
motorcycle = Motorcycle("Harley-Davidson", "Sportster")
vehicle_info(car)
vehicle_info(motorcycle)
```
