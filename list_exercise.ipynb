```python
### Assignment 8: List of Dictionaries

# Create a list of dictionaries where each dictionary represents a student with keys 'name' and 'score'. Sort the list of dictionaries by the 'score' in descending order and print the sorted list.

students = [
    {'name': 'Alice', 'score': 85},
    {'name': 'Bob', 'score': 92},
    {'name': 'Charlie', 'score': 78},
    {'name': 'David', 'score': 90}
]
sorted_students=[]
for name, score in enumerate(students):
    print(f"Student Name: {score['name']}, Score: {score['score']}")
    sorted_students = sorted(students, key = lambda x: x['score'], reverse=True)

print("Sorted Students by score in descending order:", sorted_students)
```

```python
### Assignment 10: Flattening a Nested List

nested_list = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

flatenned_list =[item for sublist in nested_list for item in sublist]
flatenned_list
```
