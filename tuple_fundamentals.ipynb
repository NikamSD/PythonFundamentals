#### Mixed tuple and corresponding operations

```python
mixed_tuple=(1,"Hello World",3.14, True)
mixed_tuple
```

```python
mixed_tuple[-2]
```

```python
mixed_tuple[::2]
```

```python
sorted_elements = tuple(sorted(mixed_tuple, key=str))
sorted_elements

###  The key=str parameter tells Python to convert each element to a string before comparing them for sorting. 
```

```python
new_sort = sorted(mixed_tuple, key=lambda x: str(x)) ##expilicit function with more flexibility
print(new_sort)
type(new_sort)  # list
```

```python
sort_by_length = sorted(mixed_tuple, key=lambda x: len(str(x)))
print(sort_by_length)
```

#### Packing and Unpacking tuple

```python
#packing
packed_tuple = 1, "Name", 5.67, True
packed_tuple
```

```python
## unpacking
a, b, c, d = packed_tuple

print(a)
print(b)
print(c)
print(d)
```

```python
### Unpacking with *
first, *middle, last = packed_tuple
print(first)
print(middle)
print(last)
```

#### Nested Tuple

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))
print(nested_tuple[1][0:2])
```

```python
print(nested_tuple[2][1])
print(nested_tuple[0][::-1]) ## reversing a tuple
```

```python
## reversing is applicable only for a tuple with a nested tuple, but the changes are not reflected in the original
nested_tuple
```

#### Iterating over tuple

```python
# Basic iteration over a nested tuple
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=' ')
    print()
```

```python
# enumerate
animals = ("cat", "dog", "bird", "fish")

for index, animal in enumerate(animals):
    print(f"Index: {index}, Animal: {animal}")
```

```python
# with range
for i in range(len(animals)):
    print(f"Index: {i}, Animal: {animals[i ]}")
```

#### Tuple Concatenation and repetition

```python
combined_tuple = animals + packed_tuple
print(combined_tuple)
```

```python
repeated_tuple = packed_tuple * 2
print(repeated_tuple)
```

```python
### Combining operations
result = (animals + packed_tuple) * 2
print(result)

new_combi = (1,) *2
print(new_combi)

additional_tuple = (1,) *2 +(5,)*3
print(additional_tuple)
```

#### Tuple Comprehension

```python
squares = tuple(x**2 for x in range(5))
squares
```

```python
##Transformation example
uppercase_animals = tuple(animal.upper() for animal in animals)
uppercase_animals
```

#### Tuple with zip() and map()

```python
zipped = zip(animals, packed_tuple)
zipped_tuple = tuple(zipped)
zipped_tuple
```

```python
for animal, info in zipped_tuple:
    print(f"Animal: {animal}, Info: {info}")
```

```python
lowercase_animals = tuple(map(lambda x: x.lower(), animals))
lowercase_animals
##map??
```
