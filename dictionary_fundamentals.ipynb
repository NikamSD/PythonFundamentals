#### Iterating over dictionaries

```python
student = {
    'name': 'Alice',
    'age': 22,
    'courses': ['Math', 'CompSci']
    }
student["country"] = "USA"
student
```

```python
##iteration generic way
for key in student:
    print(f"Key: {key}, Value: {student[key]}")
```

```python
## explicit iteration using keys
for key in student.keys():
    print(f"Key: {key}")
```

```python
## over values
for value in student.values():
    print(f"Value: {value}")    
```

```python
## key-value pair
for key, value in student.items():
    print(f"Key: {key}, Value: {value}")
```

```python
## with enumenrate for index

for idx, (key, value) in enumerate(student.items()):
    print(f"Index: {idx}, Key: {key}, Value: {value}")
```

#### Loops with Data structures

```python
## Loop through dict with list values

student_courses = {
    'Alice': ['Math', 'CompSci'],
    'Bob': ['History', 'Art'],
    'Charlie': ['Biology', 'Chemistry']
}

for name, courses in student_courses.items():
    print(f"{name} is enrolled in:")
    for course in courses:
        print(f" - {course}")
```

```python
## loop through dict with tuple values
employee_salaries = {
    'John': (50000, 'USD'),
    'Emma': (60000, 'USD'),
    'Liam': (55000, 'USD')
}

for employee_name, (salary, currency) in employee_salaries.items():
    print(f"Employee: {employee_name}, Salary: {salary} {currency}")
```

```python
## Looping through dict through set values
product_prices = {
    'Laptop': {999.99, 1099.99, 1199.99},
    'Smartphone': {499.99, 599.99, 699.99},
    'Tablet': {299.99, 399.99, 499.99}
}

for product, prices in product_prices.items():
    for price in prices:
        print(f"Product: {product}, Price: {price}")
```

```python
## Looping through dict with nested dict values
company_departments = {
    'HR': {'Manager': 'Alice', 'Employees': 5},
    'IT': {'Manager': 'Bob', 'Employees': 10},
    'Sales': {'Manager': 'Charlie', 'Employees': 7}
}

for department, employees_info in company_departments.items():
    for role, count in employees_info.items():
        print(f"Department: {department}, {role}: {count}")
```

#### unpacking 

```python
## unpacking with **operator
dict_a = {'name': 'Alice', 'age': 30}
dict_b = {'city': 'New York', 'job': 'Engineer'}

merged ={**dict_a, **dict_b}
print(merged)  # Output: {'name': 'Alice', 'age': 30, 'city': 'New York', 'job': 'Engineer'}
```

#### Dictionary Comprehension

```python
# Basic transformation
square = {i: i**2 for i in range(10)}
print(square)  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25, 6: 36, 7: 49, 8: 64, 9: 81}
```

```python
# Dictionary comprehension - from list with enumerate
fruits = ['apple', 'banana', 'cherry']
fruit_dict ={index: fruit for index, fruit in enumerate(fruits)}
print(fruit_dict)  # Output: {0: 'apple', 1: 'banana', 2: 'cherry'}
```

```python
# Dictionary Comprehension from two lists
keys = ['name', 'age', 'city']
values = ['Alice', 30, 'New York']
combined_dict = {keys[i]: values[i] for i in range(len(keys))}
print(combined_dict)  # Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}    
```

```python
#Alternative using zip
combined_dict_zip = {k: v for k, v in zip(keys, values)}
print(combined_dict_zip)  # Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}
```

```python
# Dictionary comprehension - string transformation
words = ['apple', 'banana', 'cherry']
upper_dict = {word: word.upper() for word in words}
print(upper_dict)  # Output: {'apple': 'APPLE', 'banana': 'BANANA', 'cherry': 'CHERRY'}
```

```python
## Word frequency count
text = "This is a sample text with several words this is a test text"
words = text.lower().split()
word_freq = {word: words.count(word) for word in set(words)}
print(word_freq)
```

```python
# Alternative approach
text = "This is a sample text with several words this is a test text"
words = text.split()
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1
print("Word Frequency:", frequency)
```

```python
## USe a dictionary to count he frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for num in numbers:
    if num in frequency:
        frequency[num] = frequency.get(num, 0) + 1
    else:
        frequency[num] = 1
print(frequency)  # Output: {1: 1, 2: 2, 3: 3, 4: 4}
```

```python
# Grouping data by category
students = [
    {"name": "Alice", "grade": "A"},
    {"name": "Bob", "grade": "B"},
    {"name": "Charlie", "grade": "A"},
    {"name": "Diana", "grade": "C"}
]

grouped = {}

for student in students:
    grade = student["grade"]
    
    if grade not in grouped:
        grouped[grade] = []
    grouped[grade].append(student["name"])  
print(grouped)  # Output: {'A': ['Alice', 'Charlie'], 'B': ['Bob'], 'C': ['Diana']}
```

```python
# setdefault example

graph = {}

def add_edge(source, destination):
    graph.setdefault(source, []).append(destination)

add_edge('A', 'B')
add_edge('A', 'C')
add_edge('B', 'C')        
print(graph)  # Output: {'A': ['B', 'C'], 'B': ['C']}
```

```python
# Updating the nested dictionary
employees_dict = {
    'Alice': {'age': 30, 'position': 'Manager'},
    'Bob': {'age': 25, 'position': 'Developer'},
    'Charlie': {'age': 28, 'position': 'Designer'}
}
Bob_details ={ 'Bob': {'age': 26, 'position': 'Senior Developer'} }

for key, value in Bob_details.items():
    if key in employees_dict:
        employees_dict[key].update(value)
print(employees_dict)  # Output: {'Alice': {'age': 30, 'position': 'Manager'}, 'Bob': {'age': 26, 'position': 'Senior Developer'}, 'Charlie': {'age': 28, 'position': 'Designer'}}
```
