#### Mathematical operations

```python
set1 = {101, 102, 103, 104, 105, 106, 107}
set2 = {105, 106, 107, 108, 109, 110}

union_set = set1.union(set2)
print("Union:", union_set)  # Output: {101, 102, 103, 104, 105, 106, 107, 108, 109, 110}
#Union -> Common and all elements from both sets

union_operator_set = set1 | set2        
print("Union using | operator:", union_operator_set)  # Output: {101, 102, 103, 104, 105, 106, 107, 108 , 109, 110}     
```

```python
#Intersetion -> Only common elements

set1 = {101, 102, 103, 104, 105, 106, 107}
set2 = {105, 106, 107, 108, 109, 110}

intersection_set = set1.intersection(set2)
print("Intersection:", intersection_set)  # Output: {105, 106, 107}

intersection_operator_set = set1 & set2
print("Intersection using & operator:", intersection_operator_set)  # Output: {105, 106, 107}

intersectionUpdate_set = set1.intersection_update(set2)
print("Intersection Update:", set1)  # Output: {105, 106, 107}
```

```python
# Difference -> Elements in set1 but not in set2 or unique elements in set1
set1 = {101, 102, 103, 104, 105, 106, 107}
set2 = {105, 106, 107, 108, 109, 110}

difference_set = set1.difference(set2)
print("Difference:", difference_set)  # Output: {101, 102, 103, 104}

difference_operator_set = set1 - set2
print("Difference using - operator:", difference_operator_set)  # Output: {101, 102, 103, 104}

differenceUpdate_set = set1.difference_update(set2)
print("Difference Update:", set1)  # Output: {101, 102, 103, 104}
```

```python
## Symmetric Difference -> Elements in either set1 or set2 but not in both
set1 = {101, 102, 103, 104, 105, 106, 107}
set2 = {105, 106, 107, 108, 109, 110}

symmetric_difference_set = set1.symmetric_difference(set2)
print("Symmetric Difference:", symmetric_difference_set)  # Output: {101, 102, 103, 104, 108, 109, 110}

symmetric_difference_set_operator = set1 ^ set2
print("Symmetric Difference using ^ operator:", symmetric_difference_set_operator)  # Output: {101, 102, 103, 104, 108, 109, 110}
```

#### In-Place Set Operations

```python
num_set1 ={10,20,30,40,50}
num_set2 ={40,50,60,70,80}

## In place union
num_set1_copy = num_set1.copy()
num_set1_copy |= num_set2
num_set1_copy  # Output: {10, 20, 30, 40, 50, 60, 70, 80}
```

```python
## In-place intersection
num_set1_copy = num_set1.copy()
num_set1_copy &= num_set2
num_set1_copy  # Output: {40, 50}       
```

```python
## In-place difference
num_set1_copy = num_set1.copy()
num_set1_copy -= num_set2
num_set1_copy  # Output: {10, 20, 30}
```

```python
## In-place symmetric difference
num_set1_copy = num_set1.copy()
num_set1_copy ^= num_set2
num_set1_copy  # Output: {10, 20, 30, 60, 70, 80}
```

```python
## update()
num_set1_copy = num_set1.copy()
num_set1_copy.update(num_set2)
num_set1_copy  # Output: {10, 20, 30, 40, 50, 60, 70, 80}
```

#### Set Comprehension

```python
# Transformation with set comprehension
word_set_fruits = {'apple', 'banana', 'cherry', 'date', 'apple'}
upper_fruits = {fruit.upper() for fruit in word_set_fruits}
upper_fruits  # Output: {'APPLE', 'BANANA', 'CHERRY', 'DATE'}
```

```python
word_set_fruits = {'apple', 'banana', 'cherry', 'date', 'apple', 'guava'}
fruits_length = {len(fruit) for fruit in word_set_fruits}
for fruit in word_set_fruits:
    print(f"{fruit}: {len(fruit)}")
print(fruits_length)  # Output: {4, 5, 6}
```

```python
# Complex set comprehension
result = {x*y for x in range(1, 4) for y in range(1, 4)}
print("Products:", result)
```
